```python
risk_map = {
    'Low': 1,
    'Medium': 2,
    'High': 3
}
```

```python
df['Risk_Score'] = df['Automation_Level'].map(risk_map)
```

```python
future_jobs = df[df['Job_Title'].str.strip() == 'Increasing']
```

```python
top10 = (
    future_jobs.groupby('Industry')['Risk_Score']
    .mean()
    .sort_values()
    .head(10)
)
```

```python
plt.figure(figsize=(12,6))
top10.plot(kind='bar')

plt.title('Top 10 Most Future-Proof Jobs')
plt.xlabel('Job Title')
plt.ylabel('Average Automation Risk')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
scatter = plt.scatter(
    df['Years_Experience'],
    df['Average_Salary_USD'],
    c=df['AI_Replacement_Risk'],
    alpha=0.6
)
plt.colorbar(scatter,label='AI Impact Level')
plt.title('Experience Required vs Median Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
df['AI_Impact_Level'] = pd.cut(
    df['AI_Replacement_Risk'],
    bins=[0,0.33,0.66,1],
    labels=['Low','Medium','High']
)
```

```python
education_breakdown = pd.crosstab(
    df['Education_Level'],
    df['AI_Impact_Level']
)
```

```python
education_breakdown
```

```python
ax = education_breakdown.plot(
    kind='bar',
    stacked=True,
    figsize=(12,6),
    colormap='Set2'
)
plt.title('Education Level by AI Impact Level', fontsize=14, fontweight='bold')
plt.xlabel('Education Level')
plt.ylabel('Number of Jobs')
plt.legend(title='AI Impact Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('AI_Impact_on_Jobs_2030.csv')
top_jobs = df.groupby('Job_Title')['Future_Demand_Score'].mean().sort_values(ascending=False).head(10)
industry_salary = df.groupby('Industry')['Average_Salary_USD'].mean().sort_values(ascending=False).head(10)
df['AI_Impact_Level'] = pd.cut(
    df['AI_Replacement_Risk'],
    bins=[0, 0.33, 0.66, 1],
    labels=['Low', 'Medium', 'High']
)
education_breakdown = pd.crosstab(
    df['Education_Level'],
    df['AI_Impact_Level']
)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 14))
fig.suptitle('Comprehensive Analysis of AI Impact on Jobs', fontsize=16, fontweight='bold', y=1.02)
axes[0, 0].bar(top_jobs.index, top_jobs.values, color='skyblue')
axes[0, 0].set_title('Top 10 Job Titles by Future Demand', fontsize=12)
axes[0, 0].set_xlabel('Job Title')
axes[0, 0].set_ylabel('Future Demand Score')
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 1].bar(industry_salary.index, industry_salary.values, color='lightcoral')
axes[0, 1].set_title('Top 10 Industries by Average Salary (USD)', fontsize=12)
axes[0, 1].set_xlabel('Industry')
axes[0, 1].set_ylabel('Average Salary (USD)')
axes[0, 1].tick_params(axis='x', rotation=45)
scatter = axes[1, 0].scatter(
    df['Years_Experience'],
    df['Average_Salary_USD'],
    c=df['AI_Replacement_Risk'],
    cmap='viridis',
    alpha=0.6
)
fig.colorbar(scatter, ax=axes[1, 0], label='AI Replacement Risk', orientation='vertical')
axes[1, 0].set_title('Years of Experience vs. Average Salary by AI Risk', fontsize=12)
axes[1, 0].set_xlabel('Years of Experience')
axes[1, 0].set_ylabel('Average Salary (USD)')
education_breakdown.plot(
    kind='bar',
    stacked=True,
    ax=axes[1, 1],
    colormap='Set2'
)
axes[1, 1].set_title('Education Level Distribution by AI Impact Level', fontsize=12)
axes[1, 1].set_xlabel('Education Level')
axes[1, 1].set_ylabel('Number of Employees')
axes[1, 1].legend(title='AI Impact Level', bbox_to_anchor=(1.05, 1), loc='upper left')
axes[1, 1].tick_params(axis='x', rotation=45)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()
```
